```python
df = pd.read_csv('diabetes.csv')
```

```python
flights = sns.load_dataset("flights")
```

```python
may_flights = flights.query("month == 'May'")
sns.lineplot(data=may_flights, x="year", y="passengers")
```

```python
sns.lineplot(data=df, x='Age', y='BloodPressure')
```

```python
sns.histplot(data=df, x='Age', bins=7, kde=True)
```

```python
sns.kdeplot(data=df, x='Age')
```

Criamos a coluna resultado

```python
df['Resultado'] = ''
```

```python
df.loc[df['Outcome'] == 1, 'Resultado'] = 'Diabético'
df.loc[df['Outcome'] == 0, 'Resultado'] = 'Não Diabético'
sns.countplot(data=df, x='Resultado')
```
